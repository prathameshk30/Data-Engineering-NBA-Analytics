```python
import os
from selenium import webdriver
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import regex as re
import numpy as np
import pandas as pd

DRIVER_PATH = r"C:\Fall 2022 courses\Project\chromedriver.exe"
driver = webdriver.Chrome(executable_path=DRIVER_PATH)
```

```python
element_list=[]
for i in range(1,3):
    for j in range(1,4):
        for k in range(1,6):
            element_list.append(('//*[@id="fittPageContainer"]/div[3]/div/div/div[1]/div[2]/div'+str([i])+'/div'+str([j])+'/div[2]/div'+str([k])+'/div/section/div/a/h2'))
```

```python
dataframe_list=[]
for i in range(len(element_list)):
    driver.get('https://www.espn.com/nba/')
    driver.find_element("xpath", '/html/body/div[5]/div[2]/header/nav[2]/div/ul/li[7]/a').click()
    driver.find_element("xpath",(element_list[i])).click()
    driver.find_element("xpath",'/html/body/div[1]/div/div/div/main/div[2]/div[2]/nav/ul/li[3]/a').click()
    for a in range(5,7):
        team_name=[]
        table_name=[]
        column_name=[]
        player_name=[]
        score=[]
        team_name.append(driver.find_element("xpath",('/html/body/div[1]/div/div/div/main/div[2]/div[1]/div/div/div/div/h1/span')).text)
        table_name.append(driver.find_element("xpath",'/html/body/div[1]/div/div/div/main/div[2]/div[5]/div/div/section/div/div'+str([a])+'/div[1]').text)
        q=(driver.find_elements("xpath", '/html/body/div[1]/div/div/div/main/div[2]/div[5]/div/div/section/div/div'+str([a])+'/div[2]/table/tbody/*/td/span/a'))
        for s in q:
            player_name.append(s.text)
        v=(driver.find_elements("xpath", '/html/body/div[1]/div/div/div/main/div[2]/div[5]/div/div/section/div/div'+str([a])+'/div[2]/div/div[2]/table/thead/tr/*/span/a'))
        for h in v:
            column_name.append(h.text)
        if a==5:
            for m in range (1,len(column_name)+1):
                for n in range(1, len(player_name)+1):
                    score.append(driver.find_element("xpath",'/html/body/div[1]/div/div/div/main/div[2]/div[5]/div/div/section/div/div'+str([a])+'/div[2]/div/div[2]/table/tbody/tr'+str([n])+'/td'+str([m])+'/span').text)
        else:
            for m in range (1,len(column_name)+1):
                for n in range(1,len(player_name)+1):
                    score.append(driver.find_element("xpath",'/html/body/div[1]/div/div/div/main/div[2]/div[5]/div/div/section/div/div'+str([a])+'/div[2]/div/div[2]/table/tbody/tr'+str([n])+'/td'+str([m])+'/span').text)
        score_1=np.array(score)
        if  a==5:
            score_5=score_1.reshape(len(column_name),len(player_name)).transpose()
        else:
            score_5=score_1.reshape(len(column_name),len(player_name)).transpose()
        table_name=[x for x in table_name for i in range(len(player_name))]
        team_name=[x for x in team_name for i in range(len(player_name))]
        df = pd.DataFrame(score_5, columns = column_name)
        df2= pd.DataFrame(player_name, columns = ['NAME'])
        df3= pd.DataFrame(table_name, columns = ['TABLE_TYPE'])
        df4= pd.DataFrame(team_name, columns = ['TEAM_NAME'])
        df5=pd.concat([df4,df3,df2,df], axis=1)
        print(df5)
        dataframe_list.append(df5)
```

```python
for i in dataframe_list:
    i['TEAM_NAME']=[re.sub('\\n','', str(x)) for x in i['TEAM_NAME']]
```
